# Financial timeseries data

In this little notebook we explain various ways of handling financial timeseries data. If you have a data science project involving timeseries, this notebook might help you at handling this particular kind of data. Indeed, when it comes to training machine learning models, the data should be handled decently to make the training process easy and fast. We focus here on cryptocurrency financial data.

```python
import time
import ccxt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the data.

For this task, we will use the ccxt library to connect the Binance exchange platform. There, we will load the OHLCV data for several symbols or asset pairs. While loading, we will also transform the dataset into a pandas Dataframe, more convenient to use when handling timeseries.


```python
# Set the data hyper parameters.
exchange = "binance"
symbols = ["BTC/USDT", "ETH/USDT", "ETH/BTC"]
period = "12h"
columns = ["time", "o", "h", "l", "c", "v"]
```

Let's define a little routine that converts raw data returned from the exchange platform to a decent and easily understandable pandas Dataframe.

```python
def timeseries(data, columns):
    """Returns a pandas timeseries.
    
    :param data: list of data.
    :type data: list<list<float>>.
    :param columns: the column names.
    :type columns: list<str>.
    """
    df = pd.DataFrame(data, columns=columns)
    df = df.set_index(columns[0])
    df.index = pd.to_datetime(df.index, unit="ms")
    return df
```

Now let us call the exchange through CCXT in order to load the candle sticks for the specified list of symbols. Note that you can change the list of symbols and the exchange. Please make sure the exchange supports the ccxt fetch_ohlcv method.

```python
# Initialise the dataset as a dict.
dataset = {}

# Load the exchange from ccxt.
ex = getattr(ccxt, exchange)()

# Loop on the symbols.
for symbol in symbols:
    timer = time.time()
    # Load the data.
    data = ex.fetch_ohlcv(symbol, period)
    # Convert the dataframe.
    df = timeseries(data, columns)
    # Set the dataset with the new dataframe.
    dataset[symbol] = df
    # Compute the elapsed time.
    # Display relevant information.
    elapsed = int( 1000 * ( time.time() - timer ))    
    print("Exchange Symbol {} loaded, {} [ms]".format(symbol, elapsed))
```

```python
for ex, df in dataset.items():
    plt.figure(figsize=(12, 4))
    plt.plot(df.index, df["o"], color="b")
    plt.plot(df.index, df["h"], color="r")
    plt.plot(df.index, df["l"], color="g")
    plt.gcf().autofmt_xdate()
    plt.title(ex)
    plt.show()
```
